## Power Spectral Density (PSD)
The PSD() function provided uses scipy.periodogram() to calculate the power spectral [density] of a given sound in $V^2/Hz$. The logf and logPb Boolean inputs can be used to view the frequencies and PSD in log scales respectively. Note: logPb = True will result in the PSD scale being in decibels. For analysis of noise, logf = True and logPb = True is recommended. For general PSD viewing, logf = True and logPb = False is recommended. Additionally, the relative cumulative of the PSD (not log vs log depends on logPb input) is calculated and the percentage above and below a threshold frequency (default = 32 kHz) for ultrasound categorization.

```python
# PSD(sound,logf=True,logPb=True)
```

```python
sampling_rate = 192000
PSD(sound1,logf=True,logPb=True)
```

```python
sampling_rate = 192000
PSD(sound1,logf=True,logPb=False)
```

```python
sampling_rate = samp
PSD(sound2,logf=True,logPb=True)
```

```python
sampling_rate = samp
PSD(sound2,logf=True,logPb=False)
```

## Sound Comparisons
The previous plots can be combined and compared with the compplot() function. This function can accept 1 or 2 sounds. frequency and power log scales can be toggled like before. Additionally, spectrogram "contrast" (vmin) for the left and right sounds can be varied with vmin1 and vmin2 inputs respectively. Generally, -30 is good for .wav sounds, and -100 is good for biosound .h5 files. logfs controls the frequency y-axis on the spectrograms. logfp controls the frequency x-axis on the PSD graphs. The function also outputs audio files for sound playback.

```python
# compplot(sound1,sampling_rate1,sound2=0,sampling_rate2=192000,vmin1=-100,vmin2=-100,logfs=False,logfp=True,logPb=True, s1frange=False,s2frange=False):
```
